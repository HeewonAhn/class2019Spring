input(13개의 요인), output(price에 영향을 미치는 하나의 변수) : data

parameter : 변하는 부분(여러번의 훈련을 통해서 제일 예측을 잘 하는 값을 구해야함. 따라서 제일 중요한 부분)

model fit = train / model evalutate = test / model predict = inference(상품화), 입력값 부분만 존재(train, test 단계까지는 입력값과 출력값 필요)

input layer, hidden layer, output layer : vector

hidden layer가 하나 추가되면 hidden layer로 인해 훈련되어야 할 부분 2곳 존재하게됨

parameter는 무조건 matrix(즉, weight vector가 아닌 weight matrix)

epochs의 수 = data를 몇 번 사용하는가

hidden layer와 가장 마지막 layer에만 non-linearity 부여할 수 있음(즉, input layer를 제외한 모든 layer에 non-linearity 부여할 수 있음)

activation 종류 : sigmoid, tan h, relu, softmax

sigmoid : 모든 값들을 0과 1 사이에 넣어버림 / relu : - 값은 무조건 0 처리, +에 해당하는 값은 그대로 / softmax : 넣고자 하는 layer의 총합으로 나눠줘서 그 합을 1로 만듦

A냐 B냐(binary)를 맞추는 인공지능을 만들고자 한다면, 맨 마지막 layer(output layer)에 activation으로 sigmoid를 걸어주면 됨

multiple choice의 경우엔 activation으로 softmax를 걸어주면 됨

hidden layer : sigmoid, tan h, relu 줄 수 있음

output layer : regrresion일 땐 activation 안 줌, classification일 땐 sigmoid(binary) 혹은 softmax(multiple) 줌

classification : 예를 들어 2를 모두 다 다르게 그려도 똑같이 2로 인식하는 것이 classification

model.add(Dense(1, input_shape=[13])) : 입력 13개(그림에서 실제적으로는 14개. 즉, parameter는 14개 존재), 출력 1개인 그림 생각할 수 있어야함

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

model.add(Dense(10, input_shape = [13], activation = 'relu')) : 10개의 node를 가진 1개의 hidden layer를 만듦, hidden layer에 activation을 거는 방법(즉, loss 값을 줄일 수 있는 방법)

model.add(Dense(1)) : 마지막은 무조건 1이어야함, regression이라 activation 안 줌

결과값 : Param # 140은 14x10(10은 hidden layer node의 갯수)으로, 화살표 갯수 의미

```python
model = Sequential()
model.add(Dense(10, input_shape = [13], activation = 'relu'))
model.add(Dense(1))
model.summary()
```

epochs=1000 : epochs 값을 늘릴수록 loss 값 줄어듦(따라서 loss 값을 줄이는 것이 인공지능 전문가들이 할 일)

```python
model.fit(x_train, y_train, epochs = 1000)
```

```python
model.evaluate(x_test, y_test)
```

```python
print(model.input_shape)
print(model.output_shape)
```

```python
# INPUT::
# 1. crim: per capita crime rate by town.
# 2. zn: proportion of residential land zoned for lots over 25,000 sq.ft.
# 3. indus: proportion of non-retail business acres per town.
# 4. chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
# 5. nox: nitrogen oxides concentration (parts per 10 million).
# 6. rm: average number of rooms per dwelling.
# 7. age: proportion of owner-occupied units built prior to 1940.
# 8. dis: weighted mean of distances to five Boston employment centres.
# 9. rad: index of accessibility to radial highways.
# 10. tax: full-value property-tax rate per $10,000.
# 11. ptratio: pupil-teacher ratio by town.
# 12. black: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
# 13. lstat: lower status of the population (percent).

# TARGET::
# medv: median value of owner-occupied homes in $1000s.

model.predict(np.array([0, 1, 50, 1, 0, 5, .5, 10, 1000, 100, 10, 100, 1]).reshape(1,13))
```
