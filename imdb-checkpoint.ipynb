영화 review site

input layer : 10000개의 vocabulary size. 즉, 단어의 갯수 / 10000개를 넘어가는건 unknown 처리(즉, 상위 10000개까지만 다룸) / 0~9999까지의 사전으로 정한 10000개의 단어를 one hot coding으로 넣어줌) / 10000개의 단어(x image)가 들어가면 100x10000이 되고 중간 단계도 100x128이 될 것임. BUT, hidden layer의 vector는 무조건 1이 되어야하므로 collapse시켜서 output에 하나만 나오게 해야함(즉, input to hidden, hidden to output의 weight을 훈련시키는 것)

hidden layer : 128개, layer와 node의 갯수 임의로 바꿀 수 있음

output : 마지막에 하나있는건 0(싫어요) 아니면 1(좋아요). 즉, binary로 하기 위해 하나만 줌(0.3으로 나오면 싫어한다고 볼 수 있고, 0.56으로 나오면 좋아한다고 볼 수 있음)

### Import modules

```python
# TensorFlow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

### load data 
- #### dataset (x: imdb article -> y: text type)
- #### word2idx dict (key: word; value: index)

x_train : 단어들의 번호값이 들어감

y_train : 좋아요, 싫어요가 들어감

(x_train, y_train), (x_test, y_test) : train x,y set과 test x,y set 불러옴

word2idx : 사전인데, 단어가 몇 번인지 알려줌(key 부분에 단어가 담겨있고, value 부분에 index가 담겨있음)

```python
vocab_size = 10000
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.imdb.load_data(num_words=vocab_size)
word2idx = tensorflow.keras.datasets.imdb.get_word_index()
```

### modifiy word2idx dict (to reflect x_train, x_text)

사전을 업데이트하는 과정

k : key, v : value

v+3 : 3을 더하지 않으면, 왼쪽의 word2idx는 기존의 word2idx와 변하는 것 없이 동일해질 것(즉, 모든 word2idx의 value들에 3만큼 더하는 것)

in word2idx.items() : len(word2idx)인 88588만큼 for loop 돌게 됨

2~5번째 줄 : 이 작업들이 있어야 사전에 <PAD>, <START>, <UNK>, <UNUSED>가 등록됨. 따라서 input이 다 숫자로 이루어지게됨

```python
word2idx = {k:v+3 for k, v in word2idx.items()}
word2idx['<PAD>'] = 0
word2idx['<START>'] = 1
word2idx['<UNK>'] = 2
word2idx['<UNUSED>'] = 3
```

### create idx2word: inverse of word2idx

idx2word = {v:k for k, v in word2idx.items()} : 숫자를 넣었을 때 무슨 단어인지 알려줌

```python
idx2word = {v:k for k, v in word2idx.items()}
```

1 : START로 시작

```python
x_train[0]
```

### padding x_train, x_test

모든 단어들을 자르거나 붙여서 256으로 만들어줌

```python
x_train = pad_sequences(x_train, value=word2idx['<PAD>'], padding='post', maxlen=256)
x_test = pad_sequences(x_test, value=word2idx['<PAD>'], padding='post', maxlen=256)
```

```python
len(x_train[0])
```

### create model

256을 256x10000으로 바꿔주는게 embedding(즉, 1차원에서 2차원으로 바꿔줌)

input의 256x10000(숫자의 갯수)의 matrix에서 1의 갯수는 256개, 0의 갯수는 256x10000-256개

256x10000이 10000x128의 weight과 곱해지면 hidden layer가 256x128의 matrix가 되는데, hidden layer는 무조건 matrix가 아닌 vector의 형태이어야함. 따라서 평균을 쳐서 hidden layer를 1x128의 vector로 만들어줘서 output layer를 1x1로 만들어줘야함

```python
model = Sequential()
model.add(Embedding(vocab_size, 128))
model.add(GlobalAveragePooling1D())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
Adam = tensorflow.keras.optimizers.Adam
model.compile(optimizer = Adam(lr=0.0005), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print(model.input_shape)
print(model.output_shape)
```

### train model

```python
history = model.fit(x_train, y_train, epochs=10, validation_split=0.33)
```

```python
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

### test model

```python
model.evaluate(x_test, y_test)
```
