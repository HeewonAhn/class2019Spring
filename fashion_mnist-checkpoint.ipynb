```python
# TensorFlow and tf.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
import tensorflow 
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

tensorflow.keras.datasets.fashion_mnist.load_data() : tensorfow 밑에 keras 밑에 datasets 밑에 있는 fashion_mnist load함

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.fashion_mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
x_train.shape
```

```python
x_train[0].shape
```

fashion_mnist : 0~9번까지 되어있는 10개의 category

plt.figure(figsize=(100, 100)) : 그림 크게 볼 수 있음

for i in range(100) : 100번의 loop(0~99)를 돌 거라고 내가 정함

plt.subplot(10, 10, i+1) : 맨 처음에 i에 0이 들어갈 것

plt.imshow(x_train[i], cmap=plt.cm.binary) : 맨 처음에 i에 0이 들어가면, 결과값은 Ankle boot / 따라서 y_train[0] 입력하면 결과값은 9(왜냐하면 10개의 category 중 Ankle boot는 9번째)

```python
fashion_mnist = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
                'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize=(100,100))
for i in range(100):
    plt.subplot(10,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.title(fashion_mnist[y_train[i]], fontsize=100)
plt.show()
```

img_rows, img_cols = x_train.shape[1:] : mnist처럼 'img_rows=28, img_cols=28'로 입력하는 것과 달리 x_train.shape에서 불러올 수도 있음

따라서 x_train.shape을 입력한 결과값이 (60000, 28, 28)이면, 윗 줄에서 img_rows와 img_cols의 값은 각각 28, 28

```python
img_rows, img_cols = x_train.shape[1:]

model = Sequential()
model.add(Flatten(input_shape=(img_rows, img_cols)))
model.add(Dense(512, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

lr=0.0005 : lr을 작게 설정함

```python
Adam = tensorflow.keras.optimizers.Adam
model.compile(optimizer = Adam(lr=0.0005), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
print(model.input_shape)
print(model.output_shape)
```

왜 validation이 필요한지, 왜 overfit이 되면 안되는지 알 수 있음

```python
history = model.fit(x_train, y_train, epochs=10, validation_split=0.33)
```

summaraize history for accuracy 결과값 - train : epochs 수가 많아질수록 계속 좋아지겠구나를 알 수 있음(BUT, 일반화가 안 되기 때문에 epochs 수가 많아지는게 좋다고 말할 순 없음) / test : 3 epochs쯤 훅 떨어지는게 overfit되는 상황

summarize history for loss - test : 3 epochs쯤에서 훈련되지 않은 new data에 대해서 loss가 올라감(즉, overfit되는 상황)

```python
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

test ID = 89 : test set에서 89번째를 불러옴

out : model이 predict한 값

reshape(1, img_rows, img_cols) : 원래 28x28인데, reshape으로 한 차원 더 넣어줌

```python
testID = 89
fashion_mnist = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
                'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
plt.imshow(x_test[testID], cmap=plt.cm.binary)
out = model.predict(x_test[testID].reshape(1,img_rows, img_cols))
print(fashion_mnist[np.argmax(out)])
```
